```python

def visualize_training_results(results):
    """
    Plots the loss and accuracy for the training and testing data
    """
    history = results.history
    plt.figure(figsize=(12,4))
    plt.plot(history['val_loss'])
    plt.plot(history['loss'])
    plt.legend(['val_loss', 'loss'])
    plt.title('Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.show()
    #plt.savefig("drive/MyDrive/crypto/Crypto_data/loss.png")
    
    plt.figure(figsize=(12,4))
    plt.plot(history['val_accuracy'])
    plt.plot(history['accuracy'])
    plt.legend(['val_accuracy', 'accuracy'])
    plt.title('Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.show()
    #plt.savefig("drive/MyDrive/crypto/Crypto_data/accuracy.png")
    
    
def split_sequence(seq, n_steps_in, n_steps_out):
    """
    Splits the univariate time sequence
    """
    X, y = [], []
    
    for i in range(len(seq)):
        end = i + n_steps_in
        out_end = end + n_steps_out
        
        if out_end > len(seq):
            break
        
        seq_x, seq_y = seq[i:end], seq[end:out_end]
        
        X.append(seq_x)
        y.append(seq_y)
    
    return np.array(X), np.array(y)

def split_multi_seq(seq, n_steps_in, n_steps_out):
    """
    Here seq is of the form [[prices], [volumes], ...]
    """
    X, y = [], []
    for i in range(len(seq[0])):
        seq_x, seq_y = [], []
        end = i + n_steps_in
        out_end = end + n_steps_out
              
        if out_end > len(seq[0]):
            break
        
            
        for k in range (i, end): 
            seq_x.append([seq[0][k], np.array([seq[1][k]], dtype=np.float32)])


        for k in range(end, out_end):
            seq_y.append(seq[0][k])
        y.append(seq_y)
        X.append(seq_x)
    print(X[:3])
    return np.array(X), np.array(y)



def layer_maker(n_layers, n_nodes, activation, drop=None, d_rate=.3):
    """
    Create a specified number of hidden layers for an RNN
    Optional: Adds regularization option, dropout layer to prevent potential overfitting if necessary
    """
    
    # Creating the specified number of hidden layers with the specified number of nodes
    for x in range(1,n_layers+1):
        model.add(LSTM(n_nodes, activation=activation, return_sequences=True))

        # Adds a Dropout layer after every Nth hidden layer (the 'drop' variable)
        try:
            if x % drop == 0:
                model.add(Dropout(d_rate))
        except:
            pass

```

```python
# How many periods looking back to train
n_per_in  = 30

# How many periods ahead to predict
n_per_out = 3

# Features (in this case it's 1 because there is only one feature: price)
n_features = 2
close = list(df['close'])
vol = list(df['Volume USDT'])

# Splitting the data into appropriate sequences if we only consider price
X, y = split_multi_seq([close, vol], n_per_in, n_per_out)
print(X.shape)
print(y.shape)

# Reshaping the X variable from 2D to 3D
X = X.reshape((X.shape[0], X.shape[2], n_features))
#print(X)
```

```python
# Instantiating the model
model = Sequential()

# Activation
activ = "tanh"

# Input layer
model.add(LSTM(30, activation=activ, return_sequences=True, input_shape=(n_per_in, n_features)))

# Hidden layers
layer_maker(n_layers=6, n_nodes=30, activation=activ, drop = 1)

# Final Hidden layer
model.add(LSTM(10, activation=activ))

# Output layer
model.add(Dense(n_per_out))

# Model summary
model.summary()
```

```python
class TrainingCallback(Callback):
        
    def on_epoch_end(self, epoch, logs=None):
        print("bla")
        
def mse(y_true,y_pred):
    print(type(y_true))
    S = 0
    for i in range (len(y_true)):
        S+=(y_true-y_pred)**2
    return S
```

```python
optimizer = 'adam'
loss = 'mean_squared_error'
model.compile(optimizer, loss, metrics=['accuracy'])
res = model.fit(X, y, epochs=5, batch_size=32, validation_split=0.1, verbose = 1, callbacks=[TrainingCallback()])
```

```python
def xgb_split_multi_seq(seq, n_steps_in, n_steps_out, cat):
    """
    Here seq is of the form [[prices], [volumes], ...]
    """
    X, y = [], []
    for i in range(len(seq[0])):
        seq_x, seq_y = [], []
        end = i + n_steps_in
        out_end = end + n_steps_out
              
        if out_end > len(seq[0]):
            break
        
        scaler = MinMaxScaler()
        volume = np.array(seq[1][max(i-50, 0):end]) # the max part is to have a volume over 10 days
        #print(volume)
        volume = volume.reshape(-1, 1)
        
        scaler.fit(volume) #I'm scaling the volume only over the past 30 hr or so: It doen't 
                                    #make sense to scale it over the past 5 years
        volume = scaler.transform(volume)
        
        volume = list(volume.reshape(volume.shape[0]))
        #print(volume)
        
        # This is a trick I'm trying: I can't have my training data of the for [ [0,0,1], [volume]]
        #as it poses a pb when I transform it to a tensor. I'm therefore concatenating it in the form
        #[ [0,0,1,volume] ]
        for k in range (i, end): 
            
            category = list(seq[0][k])
            category.append(volume[k-i])
            category.append(seq[2][k])
            
            seq_x.append(category)


        for k in range(end, out_end):
            seq_y.append(cat[k])
        y.append(seq_y)
        X.append(seq_x)
    #print(X[:3])
    return np.array(X), np.array(y)
```
